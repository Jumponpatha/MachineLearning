```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating random data of Normal (Gaussian) Distribution
np.random.seed(19680801)

mean = 100
sd = 15
num_points = 1000
mean2, sd2, num_points2 = 60, 10, 10000 #another sample from another class
#Get num_points points sampled from a normal distribution
samples = np.random.normal(mean, sd, num_points) #gives random num_points from G(mean, sd).

#np.random.rand samples from a uniform distribution (0,1) instead of normal distribution
#samples = num_points * np.random.rand(300) #each sample is random [0,100) or 0..99
```

```python
y_axis_scatter = np.zeros_like(samples) #zeros_like creates zeros of same data structure
plt.plot(samples, y_axis_scatter, 'o', alpha = 0.1) #plot with transparency. (x, y) = (samples, y_axis_scatter)
plt.show()
```

```python
#Graphing 1-D Gaussian
num_bins = 100 #can try 50 also.  But too many bins will discretize the normal shape

#matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None,
#  cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical',
#  rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)[source]
n, bins, patches = plt.hist(samples, num_bins, density = True, color ='green')
#n - values of the histogram bins
#bins - edges or starting points of the bins
#patches - list of individual patches used to create the histogram.
```

```python
print ("n = \n", n, "\nbins = \n", bins, "\npatches = \n", patches)
print(n.shape)
print(bins.shape)
print(patches[20]) #each patch is 1 rectangle to draw bin
```

\begin{align}
f(x,\mu,\sigma) = \frac{1}{ \sigma
 \sqrt{2 \pi}} e^{-\frac{{\left(x - \mu \right)}^{2}}{2 \sigma^{2}}} \\ \\
P(x,\mu,\sigma) = \frac{1}{Z}f(x, \mu, \sigma)
\end{align}

```python
n, bins, patches = plt.hist(samples, num_bins, density = True, color ='green')


ideal_y = ((1 / (np.sqrt(2 * np.pi) * sd)) *
     np.exp(-0.5 * (1 / sd * (bins - mean))**2))
plt.plot(bins, ideal_y, '--', color ='black') #plot (x, y) where x is bins, y is ideal_y

#set y value to
y_offset = -0.01
y_axis_scatter = np.zeros_like(samples)+ y_offset
plt.plot(samples,y_axis_scatter, 'o', color = "red", alpha = 0.1)

plt.xlabel('sampled value')
plt.ylabel('count in bin')
plt.title('matplotlib.pyplot.hist() function Example\n\n',
          fontweight ="bold")
plt.show()
```

```python
samples2 = np.random.normal(mean2, sd2, num_points2)
```

```python
#histogram plot samples and samples2

#Plot histograms for data in samples and samples 2
n, bins, patches = plt.hist(samples, num_bins, density = True, color ='green', alpha = 0.6)

n, bins2, patches = plt.hist(samples2, num_bins, density = True, color ='purple', alpha = 0.6)

#plot the ideal graphs from gaussian equation first
ideal_y = ((1 / (np.sqrt(2 * np.pi) * sd)) *
     np.exp(-0.5 * (1 / sd * (bins - mean))**2))

ideal_y2 = ((1 / (np.sqrt(2 * np.pi) * sd2)) *
     np.exp(-0.5 * (1 / sd2 * (bins2 - mean2))**2))

plt.plot(bins, ideal_y, '--', color ='black')
plt.plot(bins2, ideal_y2, '--', color ='black')


#scatter plot data in samples
y_offset = -0.01
y_axis_scatter = np.zeros_like(samples)+ y_offset
plt.plot(samples, y_axis_scatter, markersize=.2,marker = 'o', color = "red", alpha = 0.1)

#scatter plot data in samples2
y_offset2 = -0.02
y_axis_scatter = np.zeros_like(samples2)+ y_offset2
plt.plot(samples2,y_axis_scatter, markersize=.02, marker='o', color = "blue", alpha = 0.1)
plt.show()
```
