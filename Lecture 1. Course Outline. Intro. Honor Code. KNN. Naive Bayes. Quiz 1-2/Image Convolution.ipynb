```python
import cv2
import numpy as np
```

```python
#this was created because Google Colab does not allow cv2.imshow, so must patch by cv2_imshow.
#If we switch over to regular jupyter notebook not on Colab, we can change c2_imshow to cv2.imshow.
from google.colab.patches import cv2_imshow #only used when running in Google Colab
def my_imshow(title, img ):
  print(title)
  cv2_imshow(img) #should be changed to c2.imshow when not in Colab
```

```python
path = ""
fileName = path + "cat_dog_hug.jfif" #"printed.jpg" #+ #+ "Lenna.png" #  + "printed.jpg"  + "cat_dog_hug.jfif" + "Lenna.png"

#RGB images in BGR order in OpenCV
image = cv2.imread(fileName, cv2.IMREAD_COLOR)

# Print error message if image is null
if image is None:
    print('Could not read image')
else:
    print("Image file read success...")
```

```python
def myImshow(title, img):
    """
    function to make windows display work in jupyter notebook
    - shows image in a separate window,
    - waits for any key to close the window.

    """

    cv2.startWindowThread()
    my_imshow(title, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
myImshow('Original via myImshow', image)
```

```python
# Apply identity kernel
kernel1 = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]], np.float32)

#ddepth = -1 means output same depth as input; eg., 8 bit unsigned and 3 channels => 24 bits/pixel
identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)
myImshow('Identity', identity)
```

```python
#writes to file in current path
cv2.imwrite(path+'identity.jpg', identity)
```

```python
#To display image and be able to use waitKey in jupyter notebook without function like myImshow
cv2.startWindowThread()
my_imshow('Original directly', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Apply blurring kernel
kernel2 = np.ones((10,10), np.float32) / 100  #C = 1/100. Average (Mean) filter.
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel2)
my_imshow('Original', image)
my_imshow('Kernel Blur', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
